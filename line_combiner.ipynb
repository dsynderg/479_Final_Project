```python
def fileCombiner(file1, file2, outputFile, tag):
    with open(file1, 'r') as f1, open(file2, 'r') as f2, open(outputFile, 'w') as out:
        for line1, line2 in zip(f1, f2):
            out.write(tag+" "+line1.strip() + "\t" + line2.strip() + "\n")
```

```python
fileCombiner("1_4million_English_EtoS.txt", "1_4million_Spanish_EtoS.txt", "combined_1_4million_EtoS.txt", "<Spanish>")
```

```python
def tagged_combiner(file1, file2, outputFile):
    with open(file1, 'r') as f1, open(file2, 'r') as f2, open(outputFile, 'w') as out:
        for line1 in f1:
            out.write(line1.strip()+"\n")
        for line2 in f2:
            out.write(line2.strip()+"\n")
```

```python
tagged_combiner("combined_64_thousand_EtoPop.txt",'combined_64_thousand_PoptoE.txt', 'omni_64_thousand_PopandEnglish.txt')
```

```python
import re
```

```python
def combine_languages(file1, file2, outputFile):
    with open(file1, 'r') as f1, open(file2, 'r') as f2, open(outputFile, 'w') as out:
        file1dictionary = {}
        file2dictionary = {}
        for line in f1:
            line = re.sub(r'^\s*<[^>]+>\s*', '', line)
            parts = line.strip().split('\t')
            file1dictionary[parts[0]] = parts[1]
        for line in f2:
            line = re.sub(r'^\s*<[^>]+>\s*', '', line)
            parts = line.strip().split('\t')
            file2dictionary[parts[0]] = parts[1]
        for key in file1dictionary:
            if key in file2dictionary:
                out.write(f"{key}\t{file1dictionary[key]}\t{file2dictionary[key]}\n")
```
